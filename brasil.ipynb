```python
df2 = pd.read_csv('brasil.csv', sep=',')
```

```python
df2 = df2.groupby(['estado']).max()
```

```python
df2.reset_index(inplace=True)
```

```python
sns.barplot(data= df2, x='estado', y='casosAcumulado')
plt.xticks(rotation=90)
#plt.show()
plt.savefig('flask/static/images/brasil3.png')
```

```python
sns.barplot(data= df2, x='estado', y='obitosAcumulado')
plt.xticks(rotation=90)
#plt.show()
plt.savefig('flask/static/images/brasil2.png')
```

```python
AM = df1[df1['estado']=='AM']
```

```python
RO = df1[df1['estado']=='RO']
```

```python
RO = RO.groupby(['mes']).max()
```

```python
RO.reset_index(inplace=True)
```

```python
sns.barplot(data=RO, x='mes', y='casosAcumulado')
plt.xticks(rotation=90)
```

```python
AM.groupby(['mes']).max()
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(data=AM.groupby(['mes']).max(), x='data', y='casosAcumulado', label='Casos Confirmados', marker='o',)
sns.lineplot(data=AM.groupby(['mes']).max(), x='data', y='obitosAcumulado', label='Mortes', marker='o')


plt.xlabel('Meses', fontsize=9)
plt.ylabel('Casos', fontsize=9)
plt.xticks(rotation=90)
plt.savefig('flask/static/images/amapa.png')
```
